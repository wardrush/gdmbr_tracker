```python
# Create the polygon map area
import pandas as pd
import shapely.geometry as sgeom
import matplotlib.pyplot as plt
import cartopy
```

```python
# Import trail points of interest
GMDBR_points = pd.read_csv('GDMBR-2021_v2_10k-point.csv') 
trail_longs, trail_lats = GMDBR_points["Long"], GMDBR_points["Lat"]
```

```python
# Create the figure
def generate_gdmbr_map(user_long, user_lat):
    """
    Pass the user lat and long values pulled from a GPS locator API call
    to place the tracking dot each time the program is run
    
    Returns the plot axes object
    """
    fig = plt.figure()
    
    # choose the projection, the includion of state lines, and the colors
    fig, ax = plt.subplots(figsize=(12, 12), subplot_kw={"projection": cartopy.crs.PlateCarree()})
    ax.add_feature(cartopy.feature.STATES.with_scale('50m'), zorder=2, linewidth=1.5, edgecolor='k')

    # Set map for GDMBR-relevant area
    ax.set_extent([-116, -105, 31, 50], crs=cartopy.crs.PlateCarree())

    ax.set_title('GDMBR 2021 Trail')

    # turn the longs and lats into a shapely LineString
    trail_route = sgeom.LineString(zip(trail_longs, trail_lats))
    ax.add_geometries([trail_route], cartopy.crs.PlateCarree(),
                      facecolor='none', edgecolor='b')
    
    # drop the user dot onto the map in red
    plt.scatter(user_long, user_lat,
            color="red",
            s=50,
            alpha=1,
            transform=cartopy.crs.PlateCarree()) ## Important
    
    return ax
```
